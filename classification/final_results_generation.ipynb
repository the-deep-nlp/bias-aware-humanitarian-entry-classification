```python
llm_models = ["mbert", "xlm-r", "humbert"]
training_setups = ["no_debiasing", "counterfactual_debiasing"]
protected_attributes = ["gender", "country"]
visualized_methods = [
    "probability_discrepancy",
]
training_architectures = ["base_architecture", "multiabel_architecture"]

tagwise_results_df = generate_tagwise_results(
    llm_models,
    training_setups,
    training_architectures,
    protected_attributes,
    visualized_methods,
    folder_name,
    save_vizus=True,
)
```

```python
tagwise_results_df
```

```python

mapping_training_type_to_name = {
    "no_debiasing": "Base",
    "counterfactual_debiasing": "CDA",
}

mapping_training_architecture2name = {
    "multiabel_architecture": "OURS",
    "base_architecture": "Base",
}
```

```python
import pandas as pd
from tqdm import tqdm

results_df_one_run = pd.DataFrame()
# with tqdm(
#     total=len(llm_models)
#     * len(training_setups)
#     * len(protected_attributes)
#     * len(visualized_methods)
#     * len(training_architectures)
# ) as pbar:
for llm_model in llm_models:
    for training_setup in training_setups:
        for training_architecture in training_architectures:
            for bias_method in visualized_methods:
                bias_results = {}
                for bias_attribute in protected_attributes:
                    df_one_result = tagwise_results_df[
                        (tagwise_results_df.llm_model == llm_model)
                        & (tagwise_results_df.trainig_setup == training_setup)
                        & (
                            tagwise_results_df.training_architecture
                            == training_architecture
                        )
                        & (tagwise_results_df.bias_method == bias_method)
                        & (tagwise_results_df.bias_attribute == bias_attribute)
                    ]
                    #print(df_one_result.shape[0])
                    n_keywords = df_one_result["original->couterfactual"].nunique()
                    assert n_keywords == 6, f"problem, {n_keywords} total keywords found., {df_one_result['original->couterfactual'].unique()}"

                    bias_results[bias_attribute] = (
                        df_one_result.median_shift.abs().sum() / n_keywords
                    )

                metrics_one_row = {
                    "Measurement": "Probability Discepancy",
                    "LLM": llm_model,
                    "Model": f"{mapping_training_architecture2name[training_architecture]}-{mapping_training_type_to_name[training_setup]}",
                }
                metrics_one_row.update(bias_results)

                results_df_one_run = results_df_one_run.append(
                    metrics_one_row, ignore_index=True
                )

results_df_one_run[["Measurement", "LLM", "Model", "gender", "country"]]
```

```python
for i, row in results_df_one_run.round(2).iterrows():
    print(" & ".join([str(row[col]) for col in ["LLM", "Model", "gender", "country"]]))
```
